```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Make sure the path 'data/Cancer_Medical_Chatbot_Survey.csv' correctly points to the location of your CSV file. 

```python
data = pd.read_csv('data/Cancer_Medical_Chatbot_Survey.csv')

print(data.head())
print(data.shape)
print(data.columns)
```

# 1. Data Cleaning:
    - a. Rename columns for easier referencing:

```python
# Assuming you've loaded your data into a dataframe called `data`
column_rename = {
    'Which cancer types are you most interested in? (This can help tailor subsequent questions)': 'Interested_Cancer_Types',
    'Preferred Information Sources:': 'Preferred_Info_Sources',
    'Satisfaction with Online Medical Platforms - 1 (Very Dissatisfied) to 5 (Very Satisfied)': 'Platform_Satisfaction',
    'Challenges in Seeking Online Medical Info': 'Challenges',
    'Chatbot Utilization:': 'Chatbot_Use',
    'Desired Chatbot Features:': 'Desired_Features',
    'Value of Chatbot Image Analysis': 'Image_Analysis_Value',
    'Email Anonymized Medical Images \n': 'Medical_Images_Email'
}

data = data.rename(columns=column_rename)
```

    - b. Handle Missing Values:
Let's fill NaN values in categorical columns with "Not Specified" and the 'Challenges' column with "No Response".

```python
categorical_columns = ['Gender', 'Role', 'Interested_Cancer_Types', 'Preferred_Info_Sources', 'Chatbot_Use', 'Desired_Features', 'Image_Analysis_Value']
text_columns = ['Challenges']

data[categorical_columns] = data[categorical_columns].fillna("Not Specified")
data[text_columns] = data[text_columns].fillna("No Response")
```

# 2. Data Analysis:
    - a. Understand the Demographics:

```python
gender_distribution = data['Gender'].value_counts()
role_distribution = data['Role'].value_counts()
```

    - b. Analyze Chatbot Utilization and Desired Features:

```python
chatbot_use_distribution = data['Chatbot_Use'].value_counts()
desired_features_distribution = data['Desired_Features'].value_counts()
```

# 3. Data Visualization:
    - To visualize the results, you can use libraries like matplotlib and seaborn. Here's a simple bar plot for gender distribution:

```python

sns.countplot(data=data, x='Gender')
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Demographics', hue='Gender', order=data['Demographics'].value_counts().index)
plt.title('Age and Gender Distribution')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, y='Interested_Cancer_Types', order=data['Interested_Cancer_Types'].value_counts().index)
plt.title('Interest in Cancer Types')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, y='Preferred_Info_Sources', order=data['Preferred_Info_Sources'].value_counts().index)
plt.title('Preferred Information Sources')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.countplot(data=data, x='Chatbot_Use')
plt.title('Willingness to use a Chatbot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, y='Desired_Features', order=data['Desired_Features'].value_counts().index)
plt.title('Desired Features in Chatbot')
plt.show()
```
